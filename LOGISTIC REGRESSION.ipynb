# *Logistic-regression* İle Kalp Hastalığı Teşhisi yapmak

#### k-means ile Logistic regression temel farkliliklari:
1) k-means kumeleme isi yapar ve buldugu sonuc regresiyona gore daha gengelidir.
2) K-means de label ve target degerimiz taninmamistir ve yoktur
3) logistic regression ise target degeri tanimlanir ve ona gore 1 veya 0 deger seklinde sonuc verir
4) logistic regression bu veri setinde bize daha iyi bir sonuc verdi demek oluyor ki logistic regression daha mantiklidir
5) buldugumuz (accuracy score) K-means e gore daha verimli ve daha yuksektir

### Izlenecek adimlar:

1) Gerekli kütüphaneleri yüklemek

2) *Dataseti* inceleyip Analiz yapmak

3) Modeli Çekip *dataseti* modele tanıtmak

4) Modeli test etmek 

5) Çıkan sonuçlar karşılaştırarak doğruluk oranı belirlemek (*accuracy score*)

```python
# pandas tablo halinde gösteren önemli bir kütüphanedir
# numpy kısacası Numerıcal python diye geçen dizileri matris haline getiren bir kütüphanedir
# k_means modelimizdir 
# MAE (mean absolute error) testin doğruluk oranını ölçer
# train_test_split içine ayarları yaptığımız bir komuttur
# 1. adim
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# 2.adim
# bir değişkene dataseti tanımladık
df = pd.read_csv("heart.csv")
```

```python
# datasetin hakkinda bilgiler verir
df.info()
```

```python
# datasetin hakkinda bilgiler verir
df.columns.infer_objects()
```

```python
# pandas kullanılarak bir DataFrame’den belirli bir sütunun çıkarılması ve geriye kalan sütunların yeni bir değişkene atanması anlamına gelir.
# yani target ozelligi ve 1 oldugunu x den kaldirimasi gerektigi belirtir
X = df.drop("target", axis=1)
y = df["target"]
 # train_test_split() trian ve test formatinda tanimlanmasini sagliyoruz.
X_train,X_test,y_train,y_test = train_test_split(
    X,
    y,
    test_size= 0.2,
    random_state= 36
)
# stadartscaler promptu bize feature degerlerini daha tutarili ve makinenin daha net bir deger tahmini yapmasini sagliyor
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# 3.adim
# logisticregression modelini cagirip datasetimizi uygun bir formatta tanimlamak
model = LogisticRegression()
model.fit(X_train,y_train)
# Logistic regression da etiket degeri vardir ve  .fit komutuyla tanimladik
#4.adim
# modelimizi test ettik
# pd.Dataframe formati standartscaling yaptigimiz icin numpy onu dataframe formatina cevirmemiz lazim
y_pred = model.predict(X_test)
# scler .inverse_transformu ise yaptigim olceklendirmeyi geri almak anlamina geliyor
X_test_orig = scaler.inverse_transform(X_test)
result = pd.DataFrame(X_test_orig, columns=X.columns)
result["gercek"] = y_test.values
result["predict"] = y_pred

# yaptigi tahmini bir degiskene atiyoruz
# son adim modelimize verdigimiz egitimin dogruluk testini yapiyruz
acc = accuracy_score(y_test,y_pred)
print(f"Modelin yaptigi test sonucunda {acc*100}% dogruluk oraninada hesaplanmistir")

result.head(10)
```
